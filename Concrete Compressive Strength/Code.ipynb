```python
dataset = pd.read_csv('C:\Windows\concrete_data2.csv')
```

```python
discrete_features = [feature for feature in dataset.columns if feature == 'age']
```

```python
continuous_features = [feature for feature in dataset.columns if feature not in discrete_features]
```

```python
continuous_features
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
data = dataset.copy()
```

```python
# Set style
sns.set(style="whitegrid")

# Loop through each continuous feature
for feature in continuous_features:
    plt.figure(figsize=(10, 6))
    
    # Create a boxplot and violin plot for better distribution insight
    sns.violinplot(x=data[feature], y=data['concrete_compressive_strength'], color='lightgreen', inner=None)
    sns.boxplot(x=data[feature], y=data['concrete_compressive_strength'], width=0.2, color='blue', fliersize=6)
    
    # Set labels and title
    plt.xlabel(feature.capitalize(), fontsize=12)
    plt.ylabel('Concrete Compressive Strength (MPa)', fontsize=12)
    plt.title(f'{feature.capitalize()} and Concrete Compressive Strength', fontsize=14, fontweight='bold')
    
    # Show the plot
    plt.show()
```

```python
# Copy the dataset
data = dataset.copy()

# Loop through each continuous feature
for feature in continuous_features:
    plt.figure(figsize=(10, 6))
    
    # Group data by the continuous feature and calculate the median of compressive strength
    median_values = data.groupby(feature)['concrete_compressive_strength'].median()
    
    # Plot histogram with KDE
    sns.histplot(median_values, kde=True, color='skyblue', bins=20, stat='density', linewidth=2, edgecolor='black')
    
    
    # Add a vertical line to show the median compressive strength
    plt.axvline(median_values.median(), color='red', linestyle='--', label=f'Median: {median_values.median():.2f} MPa')
    
    # Set labels and title
    plt.xlabel(f'{feature.capitalize()}', fontsize=12)
    plt.ylabel('Density / Count', fontsize=12)
    plt.title(f'Distribution of {feature.capitalize()} and Concrete Compressive Strength', fontsize=14, fontweight='bold')
    
    # Add legend to the plot
    plt.legend()

    # Show the plot
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Copy the dataset
data = dataset.copy()

# Set style
sns.set(style="whitegrid")

# Loop through each continuous feature
for feature in continuous_features:
    # Apply log transformation if feature is greater than 0 (skip if 0 exists in the feature)
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        
        
        # Hexbin plot to show the density of data points
        plt.figure(figsize=(10, 6))
        plt.hexbin(data[feature], data['concrete_compressive_strength'], gridsize=50, cmap='viridis', mincnt=1)
        
        # Add color bar to represent compressive strength density
        plt.colorbar(label='Density')
        
        # Set labels and title
        plt.xlabel(f'Log of {feature.capitalize()}', fontsize=12)
        plt.ylabel('Concrete Compressive Strength (MPa)', fontsize=12)
        plt.title(f'Log Transformation of {feature.capitalize()} vs Concrete Compressive Strength (Hexbin)', fontsize=14, fontweight='bold')
        
        # Show the plot
        plt.show()
```

```python
# Copy the dataset
data = dataset.copy()

# Set seaborn style for better aesthetics
sns.set(style="whitegrid")

# Loop through each discrete feature
for feature in discrete_features:
    # Apply log transformation only if the feature does not contain 0 values
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        
        # Create a swarm plot to visualize individual points
        plt.figure(figsize=(8, 6))
        
        
        # Swarm plot
        sns.swarmplot(x=data[feature], color='purple', size=6)
        
        # Add labels and title
        plt.title(f'Swarm Plot of Log-Transformed {feature.capitalize()}', fontsize=14, fontweight='bold')
        plt.xlabel(f'Log of {feature.capitalize()}', fontsize=12)
        plt.ylabel('Values', fontsize=12)
        
        # Show the plot
        plt.show()
```

```python
X = dataset.iloc[:,0:8]
y = dataset.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.20)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
linear = LinearRegression()
linear.fit(X_train,y_train)
```

```python
linear_predict = linear.predict(X_test)
```

```python
from sklearn.metrics import r2_score
linear_score = r2_score(y_test,linear_predict)
print("Accuracy score for Linear Regression: ",linear_score*100)
```
